## Define filtering function
Before going on with analysis, let's filter the data to remove line noise and slow movement artefacts

```python
# Define the filtering functions
import scipy.signal as scisig  
def butter_bandpass(low, high, fs, order=4):
    """
    Wrapper function for the scipy butter
    :param low: Frequency to filter above
    :param high: Frequency to filter below
    :param fs: Sampling rate (hz)
    :param order: Order of filter to use (default = 4)
    :return: Numerator (b) and denominator (a) polynomials of the IIR filter
    """
    nyq = 0.5 * fs
    b, a = scisig.butter(order, [low / nyq, high / nyq], btype='band')
    return b, a

def butter_bandpass_filter(data, low, high, fs, order=4):
    """
    Filters passed data with a bandpass butter function
    :param data: data to be bandpass filtered
    :param low: Frequency to filter above
    :param high: Frequency to filter below
    :param fs: Sampling rate (hz)
    :param order: Order of filter to use (default = 4)
    :return: filtered data (and modifies original data).
    """
    b, a = butter_bandpass(low, high, fs, order=order)
    # data = data - np.mean(data)
    return scisig.filtfilt(b, a, data)
```

## Extract bands of interest
Extract beta

```python
# Filter the data
lowpass_value = 15 # Hz
highpass_value = 30 # Hz

n_samples_left = EEGdata_left.shape[0]
n_samples_right = EEGdata_right.shape[0]
n_channels = 7

EEGdata_left_beta = np.zeros((n_channels, n_samples_left))
EEGdata_right_beta = np.zeros((n_channels, n_samples_right))

# Use i+1 in .columns to skip the first column (time)
for i in range(0, n_channels):
    EEGdata_left_beta[i] = butter_bandpass_filter(EEGdata_left[EEGdata_left.columns[i+1]], lowpass_value,\
                                      highpass_value, fs, order=4)
    EEGdata_right_beta[i] = butter_bandpass_filter(EEGdata_right[EEGdata_right.columns[i+1]], lowpass_value,\
                                      highpass_value, fs, order=4)

# Transpose back into samples x channels
EEGdata_left_beta = np.transpose(EEGdata_left_beta)
EEGdata_right_beta = np.transpose(EEGdata_right_beta)

# Convert back into easy to work with dataframes
EEGdata_left_beta = pd.DataFrame(data = EEGdata_left_beta, columns=EEGdata_left.columns[1:8])
EEGdata_left_beta['Time']=EEGdata_left['Time']

EEGdata_right_beta = pd.DataFrame(data = EEGdata_right_beta, columns=EEGdata_right.columns[1:8])
EEGdata_right_beta['Time']=EEGdata_right['Time']
```

Calculate power

```python
# Beta power for left hand MI
beta_power_left = abs(scisig.hilbert(EEGdata_left_beta[EEGdata_left_beta.columns[:7]]))**2
beta_power_left = pd.DataFrame(data = beta_power_left, columns=EEGdata_left.columns[1:8])
beta_power_left['Time']=EEGdata_left['Time']

# Beta power for right hand MI
beta_power_right = abs(scisig.hilbert(EEGdata_right_beta[EEGdata_right_beta.columns[:7]]))**2
beta_power_right = pd.DataFrame(data = beta_power_right, columns=EEGdata_right.columns[1:8])
beta_power_right['Time']=EEGdata_right['Time']
```

Plot

```python
chanOfInt = 'C3'
plt.figure(figsize=(16,4))

# Find max power value to set common y-axis
ymax_plot = np.max([np.max(beta_power_left[chanOfInt]),np.max(beta_power_right[chanOfInt])])

# Left Hand MI
plt.subplot(121)
plt.plot(beta_power_left['Time'], beta_power_left[chanOfInt])
plt.xlabel('Time (seconds)')
plt.ylabel('Power')
plt.ylim([0, ymax_plot])
plt.title('Beta Power, Left Hand MI, ' + chanOfInt)

# Right Hand MI
plt.subplot(122)
plt.plot(beta_power_right['Time'], beta_power_right[chanOfInt])
plt.xlabel('Time (seconds)')
plt.ylabel('Power')
plt.ylim([0, ymax_plot])
plt.title('Beta Power, Right Hand MI, ' + chanOfInt)
plt.show()
```

```python
chanOfInt = 'C4'
plt.figure(figsize=(16,4))

# Find max power value to set common y-axis
ymax_plot = np.max([np.max(beta_power_left[chanOfInt]),np.max(beta_power_right[chanOfInt])])

# Left Hand MI
plt.subplot(121)
plt.plot(beta_power_left['Time'], beta_power_left[chanOfInt])
plt.xlabel('Time (seconds)')
plt.ylabel('Power')
plt.ylim([0, ymax_plot])
plt.title('Beta Power, Left Hand MI, ' + chanOfInt)

# Right Hand MI
plt.subplot(122)
plt.plot(beta_power_right['Time'], beta_power_right[chanOfInt])
plt.xlabel('Time (seconds)')
plt.ylabel('Power')
plt.ylim([0, ymax_plot])
plt.title('Beta Power, Right Hand MI, ' + chanOfInt)
plt.show()
```

Box Plots

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()

# Left Hand MI
bp1 = plt.boxplot([beta_power_left['C3'],
                   beta_power_left['C4']],
                  positions = [1,2],
                  widths = 0.6,
                  patch_artist=True,
                  showfliers=False)

# Right Hand MI
bp2 = plt.boxplot([beta_power_right['C3'],
                   beta_power_right['C4']],
                  positions = [3,4],
                  widths = 0.6,
                  patch_artist=True,
                  showfliers=False)

# fill with colors
colors = ['pink', 'lightblue']
for bplot in (bp1, bp2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

        
# draw temporary pink and light blue lines to draw as 
hP, = plt.plot([1,1], color='pink')
hB, = plt.plot([1,1], color='lightblue')
plt.legend((hP, hB),('C3', 'C4'))
hP.set_visible(False)
hB.set_visible(False)

plt.title('Beta Power, C3 and C4')
plt.ylabel('Spectral Power')
plt.xlim(0,5)
ax.set_xticklabels(['Left Hand MI','Right Hand MI'])
ax.set_xticks([1.5, 3.5])
plt.show()
```

What we expected:
- Greater spectral power for right hand MI in C3 than for in the left hand (AKA right pink larger than left pink) **<-- yes we see this**
- Greater spectral power for left hand MI in C4 than for in the right hand (AKA left blue larger than right blue) **<-- we do NOT see this**
