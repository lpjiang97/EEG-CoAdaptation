## Define filtering function
Before going on with analysis, let's filter the data to remove line noise and slow movement artefacts

```python
# Define the filtering functions
import scipy.signal as scisig  
def butter_bandpass(low, high, fs, order=4):
    """
    Wrapper function for the scipy butter
    :param low: Frequency to filter above
    :param high: Frequency to filter below
    :param fs: Sampling rate (hz)
    :param order: Order of filter to use (default = 4)
    :return: Numerator (b) and denominator (a) polynomials of the IIR filter
    """
    nyq = 0.5 * fs
    b, a = scisig.butter(order, [low / nyq, high / nyq], btype='band')
    return b, a

def butter_bandpass_filter(data, low, high, fs, order=4):
    """
    Filters passed data with a bandpass butter function
    :param data: data to be bandpass filtered
    :param low: Frequency to filter above
    :param high: Frequency to filter below
    :param fs: Sampling rate (hz)
    :param order: Order of filter to use (default = 4)
    :return: filtered data (and modifies original data).
    """
    b, a = butter_bandpass(low, high, fs, order=order)
    # data = data - np.mean(data)
    return scisig.filtfilt(b, a, data)
```

## Extract bands of interest
Extract alpha

```python
# Filter the data
lowpass_value = 8 # Hz
highpass_value = 12 # Hz

n_samples_eyes_closed = EEGdata_eyes_closed.shape[0]
n_samples_eyes_open = EEGdata_eyes_open.shape[0]
n_channels = 7

EEGdata_eyes_closed_alpha = np.zeros((n_channels, n_samples_eyes_closed))
EEGdata_eyes_open_alpha = np.zeros((n_channels, n_samples_eyes_open))

# Use i+1 in .columns to skip the first column (time)
for i in range(0, n_channels):
    EEGdata_eyes_closed_alpha[i] = butter_bandpass_filter(EEGdata_eyes_closed[EEGdata_eyes_closed.columns[i+1]], lowpass_value,\
                                      highpass_value, fs, order=4)
    EEGdata_eyes_open_alpha[i] = butter_bandpass_filter(EEGdata_eyes_open[EEGdata_eyes_open.columns[i+1]], lowpass_value,\
                                      highpass_value, fs, order=4)

# Transpose back into samples x channels
EEGdata_eyes_closed_alpha = np.transpose(EEGdata_eyes_closed_alpha)
EEGdata_eyes_open_alpha = np.transpose(EEGdata_eyes_open_alpha)

# Convert back into easy to work with dataframes
EEGdata_eyes_closed_alpha = pd.DataFrame(data = EEGdata_eyes_closed_alpha, columns=EEGdata_eyes_closed.columns[1:8])
EEGdata_eyes_closed_alpha['Time']=EEGdata_eyes_closed['Time']

EEGdata_eyes_open_alpha = pd.DataFrame(data = EEGdata_eyes_open_alpha, columns=EEGdata_eyes_open.columns[1:8])
EEGdata_eyes_open_alpha['Time']=EEGdata_eyes_open['Time']
```

Calculate power

```python
# Alpha power for eyes closed condition
alpha_power_eyes_closed = abs(scisig.hilbert(EEGdata_eyes_closed_alpha[EEGdata_eyes_closed_alpha.columns[:7]]))**2
alpha_power_eyes_closed = pd.DataFrame(data = alpha_power_eyes_closed, columns=EEGdata_eyes_closed.columns[1:8])
alpha_power_eyes_closed['Time']=EEGdata_eyes_closed['Time']

# Alpha power for eyes open condition
alpha_power_eyes_open = abs(scisig.hilbert(EEGdata_eyes_open_alpha[EEGdata_eyes_open_alpha.columns[:7]]))**2
alpha_power_eyes_open = pd.DataFrame(data = alpha_power_eyes_open, columns=EEGdata_eyes_open.columns[1:8])
alpha_power_eyes_open['Time']=EEGdata_eyes_open['Time']
```

Plot

```python
chanOfInt = 'PO7'
plt.figure(figsize=(16,4))

# Find max power value to set common y-axis
ymax_plot = np.max([np.max(alpha_power_eyes_closed[chanOfInt]),np.max(alpha_power_eyes_open[chanOfInt])])

# Eyes Closed
plt.subplot(121)
plt.plot(alpha_power_eyes_closed['Time'], alpha_power_eyes_closed[chanOfInt])
plt.xlabel('Time (seconds)')
plt.ylabel('Power')
plt.ylim([0, ymax_plot])
plt.title('Alpha Power, Eyes Closed, ' + chanOfInt)

# Eyes Open
plt.subplot(122)
plt.plot(alpha_power_eyes_open['Time'], alpha_power_eyes_open[chanOfInt])
plt.xlabel('Time (seconds)')
plt.ylabel('Power')
plt.ylim([0, ymax_plot])
plt.title('Alpha Power, Eyes Open, ' + chanOfInt)
plt.show()
```
