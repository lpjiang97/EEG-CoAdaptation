## Define filtering function
Before going on with analysis, let's filter the data to remove line noise and slow movement artefacts

```python
# Define the filtering functions
import scipy.signal as scisig  
def butter_bandpass(low, high, fs, order=4):
    """
    Wrapper function for the scipy butter
    :param low: Frequency to filter above
    :param high: Frequency to filter below
    :param fs: Sampling rate (hz)
    :param order: Order of filter to use (default = 4)
    :return: Numerator (b) and denominator (a) polynomials of the IIR filter
    """
    nyq = 0.5 * fs
    b, a = scisig.butter(order, [low / nyq, high / nyq], btype='band')
    return b, a

def butter_bandpass_filter(data, low, high, fs, order=4):
    """
    Filters passed data with a bandpass butter function
    :param data: data to be bandpass filtered
    :param low: Frequency to filter above
    :param high: Frequency to filter below
    :param fs: Sampling rate (hz)
    :param order: Order of filter to use (default = 4)
    :return: filtered data (and modifies original data).
    """
    b, a = butter_bandpass(low, high, fs, order=order)
    # data = data - np.mean(data)
    return scisig.filtfilt(b, a, data)
```

## Extract bands of interest
Extract 15 Hz

```python
# Filter the data
lowpass_value = 14.75 # Hz
highpass_value = 15.25 # Hz

n_samples_left_15 = EEGdata_left.shape[0]
n_samples_right_15 = EEGdata_right.shape[0]
n_channels = 7

EEGdata_left_15 = np.zeros((n_channels, n_samples_left_15))
EEGdata_right_15 = np.zeros((n_channels, n_samples_right_15))

# Use i+1 in .columns to skip the first column (time)
for i in range(0, n_channels):
    EEGdata_left_15[i] = butter_bandpass_filter(EEGdata_left[EEGdata_left.columns[i+1]], lowpass_value,\
                                      highpass_value, fs, order=4)
    EEGdata_right_15[i] = butter_bandpass_filter(EEGdata_right[EEGdata_right.columns[i+1]], lowpass_value,\
                                      highpass_value, fs, order=4)

# Transpose back into samples x channels
EEGdata_left_15 = np.transpose(EEGdata_left_15)
EEGdata_right_15 = np.transpose(EEGdata_right_15)

# Convert back into easy to work with dataframes
EEGdata_left_15 = pd.DataFrame(data = EEGdata_left_15, columns=EEGdata_left.columns[1:8])
EEGdata_left_15['Time']=EEGdata_left['Time']

EEGdata_right_15 = pd.DataFrame(data = EEGdata_right_15, columns=EEGdata_right.columns[1:8])
EEGdata_right_15['Time']=EEGdata_right['Time']
```

### Power at 15 Hz
Calculate power for 15 Hz

```python
# 15 Hz EEG power for looking right data file (15 Hz LED)
power_15_left = abs(scisig.hilbert(EEGdata_left_15[EEGdata_left_15.columns[:7]]))**2
power_15_left = pd.DataFrame(data = power_15_left, columns=EEGdata_left.columns[1:8])
power_15_left['Time']=EEGdata_left['Time']

# 15 Hz EEG power for looking left data file (17 Hz LED)
power_15_right = abs(scisig.hilbert(EEGdata_right_15[EEGdata_right_15.columns[:7]]))**2
power_15_right = pd.DataFrame(data = power_15_right, columns=EEGdata_right.columns[1:8])
power_15_right['Time']=EEGdata_right['Time']
```

Plot
